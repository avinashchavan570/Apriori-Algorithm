```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
price=pd.read_csv("Price.csv")
price.head()
```

```python
invoice=pd.read_csv("Invoice.csv")
invoice.head()
```

```python
invoice_details=pd.read_csv("InvoiceDetails.csv")
invoice_details.head()
```

```python
price_new = price.drop(['Brand','Category_Level_0','Category_Level_1','Category_Level_2','Category_Level_3','Fiyat'],axis=1)
```

```python
price_new.head()
```

```python
df = pd.merge(invoice_details,price_new,on='ProductID')
```

```python
df.head()
```

```python
basket = (df.groupby(['InvoiceID', 'ProductName'])['Adet']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceID'))
```

```python
basket.head()
```

```python
def encode_unit(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
```

```python
set_basket = basket.applymap(encode_unit)
set_basket.head()
```

# set parameters for apriori algorithms

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_items = apriori(set_basket, min_support=0.15,use_colnames=True)
frequent_items
```
